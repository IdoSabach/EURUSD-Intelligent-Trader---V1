```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/results.csv')
df
```

```python
df.columns
```

```python
df['long_group'] = (df['long_entry'] == 1).cumsum()
df['short_group'] = (df['short_entry'] == 1).cumsum()
```

```python
df[['long_group','short_group']]
```

```python
df['trade_id'] = df['long_group'] + df['short_group']
df['trade_type'] = np.where(df['long_entry'] == 1, 'Long', np.where(df['short_entry'] == 1, 'Short', None))
```

```python
df[['long_group','short_group','trade_id','trade_type']]
```

```python
df['trade_type'].value_counts()
```

```python
trades_df = df[df['trade_id'] > 0].copy()
```

```python
trades_df
```

```python
df['trade_id'].value_counts()
```

```python
grouped_trades = trades_df.groupby('trade_id')
grouped_trades
```

```python
grouped_trades.head()
```

```python
# יצירת טבלת סיכום עסקאות
trades_summary_df = grouped_trades.agg(
    # 1. מדדי זיהוי
    trade_type=('trade_type', 'first'),
    
    # 2. מדדי זמן
    entry_time=('Datetime', 'first'),
    exit_time=('Datetime', 'last'),
    
    # 3. מדדי מחיר
    entry_price=('price', 'first'),
    exit_price=('price', 'last')
)

# מעלים את מדד ה-trade_id לעמודה רגילה (Reset the index)
trades_summary_df = trades_summary_df.reset_index()

```

```python
# חישוב P&L באחוזים (שימוש ב np.where כדי לטפל ב-Long וב-Short)
trades_summary_df['pnl_percent'] = np.where(
    trades_summary_df['trade_type'] == 'Long',
    (trades_summary_df['exit_price'] - trades_summary_df['entry_price']) / trades_summary_df['entry_price'] * 100,
    (trades_summary_df['entry_price'] - trades_summary_df['exit_price']) / trades_summary_df['entry_price'] * 100
)

# עמודה בוליאנית פשוטה כדי לזהות זכייה
trades_summary_df['winning_trade'] = trades_summary_df['pnl_percent'] > 0
```

```python
# הגדר את גודל הפוזיציה (התאם את זה לנתונים שלך)
# אם זה קבוע, הגדר פה, לדוגמה:
POSITION_SIZE = 10  # נניח שקנית/מכרת 10 יחידות בכל עסקה

# חישוב P&L כספי
trades_summary_df['pnl_usd'] = np.where(
    trades_summary_df['trade_type'] == 'Long',
    (trades_summary_df['exit_price'] - trades_summary_df['entry_price']) * POSITION_SIZE,
    (trades_summary_df['entry_price'] - trades_summary_df['exit_price']) * POSITION_SIZE
)
```

```python
# 1. המרת עמודות הזמן לפורמט Datetime (חובה!)
trades_summary_df['entry_time'] = pd.to_datetime(trades_summary_df['entry_time'])
trades_summary_df['exit_time'] = pd.to_datetime(trades_summary_df['exit_time'])

# 2. חישוב משך העסקה (Duration) - הקוד המקורי שלך:
trades_summary_df['duration'] = trades_summary_df['exit_time'] - trades_summary_df['entry_time']

# 3. הוספת עמודת משך בדקות (להצגה נוחה)
trades_summary_df['duration_mins'] = trades_summary_df['duration'].dt.total_seconds() / 60
```

```python
def calculate_hedge_fund_metrics(df, trades_df):
    # --- תיקון: ווידוא שהאינדקס הוא תאריך ---
    df = df.copy() # עובדים על עותק כדי לא להרוס את המקורי
    
    # אם האינדקס הוא לא תאריך, ננסה להגדיר את עמודת Datetime כאינדקס
    if not isinstance(df.index, pd.DatetimeIndex):
        if 'Datetime' in df.columns:
            df['Datetime'] = pd.to_datetime(df['Datetime']) # המרה ליתר ביטחון
            df = df.set_index('Datetime')
        else:
            print("Warning: Could not find 'Datetime' column for index. Sharpe Ratio might be wrong.")

    metrics = {}
    
    # --- 1. מדדי עסקאות (Trade Statistics) ---
    total_trades = len(trades_df)
    winning_trades = trades_df[trades_df['pnl_usd'] > 0]
    losing_trades = trades_df[trades_df['pnl_usd'] <= 0]
    
    win_rate = (len(winning_trades) / total_trades) * 100 if total_trades > 0 else 0
    
    avg_win = winning_trades['pnl_usd'].mean() if not winning_trades.empty else 0
    avg_loss = losing_trades['pnl_usd'].mean() if not losing_trades.empty else 0
    
    # Profit Factor
    gross_profit = winning_trades['pnl_usd'].sum()
    gross_loss = abs(losing_trades['pnl_usd'].sum())
    profit_factor = gross_profit / gross_loss if gross_loss != 0 else np.inf

    # Expectancy
    if total_trades > 0:
        expectancy = (len(winning_trades)/total_trades * avg_win) - (len(losing_trades)/total_trades * abs(avg_loss))
    else:
        expectancy = 0

    # --- 2. מדדי סיכון ותשואה (Risk & Return) ---
    cumulative_returns = df['cstrategy']
    running_max = cumulative_returns.cummax()
    drawdown = (cumulative_returns - running_max) / running_max
    max_drawdown = drawdown.min() * 100 

    total_return = (cumulative_returns.iloc[-1] - 1) * 100

    # Sharpe Ratio (תיקון: כעת האינדקס תקין וה resample יעבוד)
    try:
        daily_returns = df['cstrategy'].resample('D').last().pct_change().dropna()
        risk_free_rate = 0.0 
        
        if daily_returns.std() != 0:
            sharpe_ratio = (daily_returns.mean() - risk_free_rate) / daily_returns.std() * np.sqrt(252)
        else:
            sharpe_ratio = 0
    except Exception as e:
        print(f"Could not calculate Sharpe Ratio: {e}")
        sharpe_ratio = 0

    # --- 3. ארגון התוצאות ---
    metrics = {
        'Total Trades': total_trades,
        'Win Rate (%)': round(win_rate, 2),
        'Profit Factor': round(profit_factor, 2),
        'Avg Win ($)': round(avg_win, 2),
        'Avg Loss ($)': round(avg_loss, 2),
        'Expectancy ($)': round(expectancy, 2),
        'Total Return (%)': round(total_return, 2),
        'Max Drawdown (%)': round(max_drawdown, 2),
        'Sharpe Ratio': round(sharpe_ratio, 2)
    }
    
    return metrics
```

```python
# וודא שוב ש-trades_summary_df חושב נכון לפי השלבים הקודמים
metrics = calculate_hedge_fund_metrics(df,trades_summary_df)

print("==== Hedge Fund Performance Report ====")
for k, v in metrics.items():
    print(f"{k:<20}: {v}")
```
